# データの加工

### １.１データの呼び出し(rainfall)

```python
dir_name ="rainfall_data"
df = []
for i in range(20):
    df_v = pd.read_csv(dir_name+"/data-"+str(i).zfill(2)+".csv",header=3,names=["time","rainfall"],encoding='cp932',usecols=[0,1],index_col="time",
				parse_dates=True)
    df.append(df_v)

df_total = pd.concat(df,axis=0)
df_total
```

### １.２データの呼び出し(temp)

```python
dir_name ="temp_data"
df = []
for i in range(20):
    df_v = pd.read_csv(dir_name+"/data-"+str(i).zfill(2)+".csv",header=2,names=["time","temp"],encoding='cp932',usecols=[0,1],index_col="time",
				parse_dates=True)
    df.append(df_v)
    
df_total_temp = pd.concat(df,axis=0)
df_total_temp
```
